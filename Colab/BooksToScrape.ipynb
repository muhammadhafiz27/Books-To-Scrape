# **Import Data**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/muhammadhafiz27/Books-To-Scrape/refs/heads/main/books_to_scrape_tugas.csv"
df = pd.read_csv(url)
display(df.head())
```

```python
df_cleaned = df.copy()
for col in ['Judul', 'harga']:
    df_cleaned[col] = df_cleaned[col].str.split('\n\n')
df_cleaned['rating'] = df_cleaned['rating'].str.split('\n')
df_cleaned = df_cleaned.explode(['Judul', 'harga', 'rating'])
display(df_cleaned.head(40))
```

# **Analisis Sederhana**

**Rata-Rata Harga Buku**

```python
df_cleaned['harga_numeric'] = df_cleaned['harga'].str.replace('£', '').astype(float)
average_price = df_cleaned['harga_numeric'].mean()
print(f"The average price of the books is £{average_price:.2f}")
```

**Buku dengan Rating “Five”**

```python
five_star_books = df_cleaned[df_cleaned['rating'] == 'star-rating Five']
num_five_star_books = len(five_star_books)
print(f"There are {num_five_star_books} books with a 'Five' rating.")
```

# **Task**
Analisis data buku dari "https://raw.githubusercontent.com/muhammadhafiz27/Books-To-Scrape/refs/heads/main/books_to_scrape_tugas.csv" untuk menghitung rata-rata harga buku dan jumlah buku dengan rating "Five", lalu berikan langkah-langkah pengerjaan dalam bahasa Indonesia.

## Muat data

### Subtask:
Muat data buku dari URL yang disediakan ke dalam pandas DataFrame.


**Reasoning**:
Load the data into a pandas DataFrame and display the head.



```python
url = "https://raw.githubusercontent.com/muhammadhafiz27/Books-To-Scrape/refs/heads/main/books_to_scrape_tugas.csv"
df = pd.read_csv(url)
display(df.head())
```

## Bersihkan data

### Subtask:
Proses data yang dimuat dengan memecah dan meledakkan kolom 'Judul', 'harga', dan 'rating' untuk memastikan setiap baris mewakili satu buku dengan detail yang sesuai.


**Reasoning**:
Process the loaded data by splitting and exploding the 'Judul', 'harga', and 'rating' columns to ensure each row represents a single book with its corresponding details.



```python
df_cleaned = df.copy()

for col in ['Judul', 'harga']:
    df_cleaned[col] = df_cleaned[col].str.split('\n\n')

df_cleaned['rating'] = df_cleaned['rating'].str.split('\n')

df_cleaned = df_cleaned.explode(['Judul', 'harga', 'rating'])

display(df_cleaned.head(40))
```

## Hitung Rata-Rata Harga Buku

### Subtask:
Hitung rata-rata harga buku dari kolom 'harga' yang telah dibersihkan.

**Reasoning**:
To calculate the average price, I need to convert the 'harga' column to a numeric format by removing the currency symbol and converting the type. Then I can calculate the average.

```python
df_cleaned['harga_numeric'] = df_cleaned['harga'].str.replace('£', '').astype(float)

average_price = df_cleaned['harga_numeric'].mean()

print(f"The average price of the books is £{average_price:.2f}")
```

## Hitung Buku dengan Rating “Five”

### Subtask:
Hitung jumlah buku dengan rating "Five" dari kolom 'rating' yang telah dibersihkan.

**Reasoning**:
To find the number of books with a "Five" rating, I will filter the DataFrame to select only those rows where the rating is 'star-rating Five' and then count the number of rows.

```python
five_star_books = df_cleaned[df_cleaned['rating'] == 'star-rating Five']

num_five_star_books = len(five_star_books)

print(f"There are {num_five_star_books} books with a 'Five' rating.")
```
